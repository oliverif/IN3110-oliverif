```python

sum_cals=[]
temp_sum=0
for c in open("input.txt").read().splitlines():
    if c =='':
        sum_cals.append(temp_sum)
        temp_sum=0
        continue
    calories = int(c)
    temp_sum +=calories
sum_cals.append(temp_sum)
sum(sorted(sum_cals)[-3:])
```
